```python
## calcuating the sum of confirmed cases, deaths and recovered persons respectively

sum_of_confirmed_cases = 0
sum_of_deaths = 0
sum_of_recovered_cases = 0
sum_of_cases = []

with open('covid_19_data.csv','r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        sum_of_confirmed_cases += float(row['Confirmed'])
        sum_of_deaths += float(row['Deaths'])
        sum_of_recovered_cases += float(row['Recovered'])

sum_of_cases.append(sum_of_confirmed_cases)
sum_of_cases.append(sum_of_deaths)
sum_of_cases.append(sum_of_recovered_cases)
print(sum_of_confirmed_cases)
print(sum_of_deaths)
print(sum_of_recovered_cases)
print(sum_of_cases)
cases = ['Confirmed', 'Deaths', 'Recovered']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Rates of cases

```python
with open('covid_19_data.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    reported_cases = {'Confirmed':[], 'Deaths':[], 'Recovered':[]}
    
    for row in csv_reader:
        reported_cases['Confirmed'].append(float(row['Confirmed']))
        reported_cases['Deaths'].append(float(row['Deaths']))
        reported_cases['Recovered'].append(float(row['Recovered']))

print(reported_cases['Confirmed'][:5])
```

### Death rates

```python
plt.plot(range(1, len(reported_cases['Deaths'])+1), reported_cases['Deaths'], '-.')
plt.title('Death Rates')
plt.xlabel('Number of times')
plt.ylabel('Deaths')
```

### Rate of confirmed cases

```python
plt.plot(range(1, len(reported_cases['Confirmed'])+1), reported_cases['Confirmed'], '-*')
plt.title('Rate of Confirmed Cases')
plt.xlabel('Number of times')
plt.ylabel('Confirmed cases')
```

### Recovery rate

```python
plt.plot(range(1, len(reported_cases['Recovered'])+1), reported_cases['Recovered'], '.')
plt.title('Rate of Recovery')
plt.xlabel('Number of times')
plt.ylabel('Recoveries')
```

```python
plt.bar(cases, sum_of_cases)
plt.title('Number of Deaths, Confirmed cases and Recoveries')
plt.xlabel('Case')
plt.ylabel('Sum of cases')
plt.style.use('bmh')
```
