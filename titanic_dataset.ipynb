## The dataset being worked on is the titanic dataset. There were 577 males and 314 females on the ship. 109 out of the 577 males survived while 233 out of the 314 females survived. Although the original number of males onboard were more than the females, the number of females who survived were more than the males. 

```python
import csv
```

```python
import csv
with open('train_titanic.csv', 'r') as titanic:
    csv_dict_reader = csv.DictReader(titanic)
    male_survived = {'Yes':0, 'No':0}
    female_survived = {'Yes':0, 'No':0}
    for row in csv_dict_reader:
        if row['Sex']=='male' and row['Survived']=='1':
            male_survived['Yes'] += 1
        elif row['Sex']=='male' and row['Survived']=='0':
            male_survived['No'] += 1
        elif row['Sex']=='female' and row['Survived']=='1':
            female_survived['Yes'] += 1
        elif row['Sex']=='female' and row['Survived']=='0':
            female_survived['No'] += 1
            
print('Male:', male_survived)
print('Female:', female_survived)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(male_survived.keys(), male_survived.values(), label = 'Humans')
plt.title('Survival of males on the titanic')
plt.style.use('seaborn-deep')
plt.ylabel('Number of people')
plt.grid(True)
plt.yticks([0,50,100,150,200,250,300,350,400,450,450], rotation=45)
plt.xticks(rotation=45)
plt.legend()
```

```python
plt.bar(male_survived.keys(), [100*x/sum(male_survived.values()) 
                               for x in male_survived.values()])
plt.title('Survival of males on the titanic')
plt.style.use('seaborn-notebook')
plt.ylabel('Number of people')
plt.yticks(rotation=30)
plt.xticks(rotation=45)
```

```python
plt.bar(male_survived.keys(), male_survived.values(), label='Male')
plt.bar(female_survived.keys(), female_survived.values(), label='Female')
plt.title('Survival of people on titanic')
plt.legend()
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.style.use('seaborn-darkgrid')
```
